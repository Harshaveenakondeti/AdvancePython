```python
url1='https://realpython.github.io/fake-jobs/'
```

```python
page1=requests.get(url1) # to webscrape page
```

```python
soup=BeautifulSoup(page1.content,'html.parser') # parse html
```

```python
results = soup.find(id='ResultsContainer')  # find by id
```

```python
print(results.prettify())
```

```python
res=
```

```python
print(r)
```

```python
all_links=soup.find_all('a') # all anchor tags
```

```python
for i in all_links:
    print(i.get('href'))  # we will get all links
```

## Find Elements by html class

```python
job_elements = results.find_all("div", class_='card-content')
```

```python
for i in job_elements:  # print in html
    title_element=i.find('h2',class_='title')
    company_element=i.find('h3',class_='company')
    location_element=i.find('p',class_='location')
    print(title_element)
    print(company_element)
    print(location_element)# extra whitespace we will get
```

```python
for i in job_elements:
    title_element=i.find('h2',class_='title')
    company_element=i.find('h3',class_='company')
    location_element=i.find('p',class_='location')
    print(title_element.text)
    print(company_element.text)
    print(location_element.text)# whitespace will present
```

```python
for i in job_elements:
    title_element=i.find('h2',class_='title')
    company_element=i.find('h3',class_='company')
    location_element=i.find('p',class_='location')
    print('title: ',title_element.text.strip())
    print('company name: ',company_element.text.strip())
    print('location: ',location_element.text.strip())# we dont have whitespace
    print()
```

```python
python_jobs=results.find_all('h2',string_='Python') # we are matching exact 
print(python_jobs)  # get elements by string
```

```python
# to overcome this , we have new concep
# pass a function to beautifulsoup method
python_job=results.find_all("h2",string=lambda text:"python" in text.lower())
# we are passing ananymos function
for i in python_job:
    print(i.text)
```
