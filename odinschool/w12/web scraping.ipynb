```python

import requests
from bs4 import BeautifulSoup
url = 'https://realpython.github.io/fake-jobs/'
page = requests.get(url)#response we got
soup = BeautifulSoup(page.content, 'html.parser')
# soup object , we can use xml also
rows = soup.find_all()
print(rows)
```

```python
page.status_code
```

```python
print(page)  # response header
```

```python
page.text # all html that it has been extracted
```

```python
results= soup.find('div') # one occurance,1st div
print(results)
```

```python
results= soup.find_all('div') #all div
print(results)
```

```python
results= soup.find('div',id="ResultsContainer") # one occurance,1st div
print(results)
```

```python
elements=results.find_all("div",class_="card-content")
print(elements)
```

```python
for job in elements:
    title=job.find("h2",class_="title")
    company=job.find("h3")
    print(title)
    print(company)
```

```python
for job in elements:
    title=job.find("h2",class_="title")
    company=job.find("h3")
    print(title.text.strip())
    print(company.text)
```

```python
# find all h2 tags
elements=results.find_all('h2')
elements
```

```python
for h in elements:
    print(h,"\n"*3)
```

```python
elements=results.find_all('h2',string="Product manager")
for h in elements:
    print(h,"\n"*3)
```
